```python
"""
A example for GFO. This example shows GFO working with synthetic features
and noise added to the resulting vector to see if it is still able to find
the relationships.
"""
from learner import GFO
from matplotlib import pyplot as plt

import numpy as np
import pandas as pd

# Generate synthetic data
A = np.matrix(np.random.randn(1000,5))

# Generate a known relationship between the synthetic
# features of A and add noise to the vector
b = 1*A[:, 0] - 2*A[:, 4]
# b += np.random.normal(b.shape[1])

"""
The answer is
     [[1]
      [0]
 b =  [0]
      [0]
      [-0.5]
      [0]
      [0]
      [0]
      [0]
      [0]]
"""


# Use GFO to attempt to learn about how the features in A determine
# the value of the answers in b.
x, r, steps = GFO(A, b).learn(lrs=20)
```

```python
"""
We round to 1 decimal place because we know the precision of the answer
only goes out 1 decimal place.
"""
x
```

```python
"""
As you can see even with noise and a lot of worthless features the method employed in
GFO still manages to find the solution to this synthetic problem. We can visualize
the convergence by plotting r.
"""
plt.figure(figsize=(22,9))
plt.plot(r)
plt.title('Convergence')
plt.xlabel('epoch')
plt.ylabel('residual')
plt.grid()
```
