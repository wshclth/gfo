```python
from matplotlib.animation import FuncAnimation
from learner import GFO
from matplotlib import pyplot as plt
from IPython.display import HTML
from matplotlib import animation, rc

import pandas as pd
import numpy as np

data = pd.read_csv('winequality-red.csv')

# Setup the A matrix (feature_set) and the b vector
feature_set = np.matrix(data.loc[:, 'fixed acidity':'alcohol'])
b = np.matrix(data.loc[:, 'quality']).T

# Learn about how these features impact the value of b
w, r, steps = GFO(feature_set, b).learn()

# Setup plot and animate
plt.style.use('seaborn-deep')
plt.rcParams["animation.html"] = "jshtml"

fig, ax = plt.subplots(figsize=(22, 5))
line = ax.plot(np.rint(feature_set*steps[0]), label='fit', alpha=0.6)[0]
bl = ax.plot(b, label='wanted')
ax.set_ylim([min(b)-2, max(b)+2])

def animate(i):
    line.set_ydata(np.rint(feature_set*steps[i]))

anim = FuncAnimation(fig, animate, interval=200, frames=11)

plt.grid()
plt.legend(loc=1)
plt.close()
anim
```
